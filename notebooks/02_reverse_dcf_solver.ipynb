# 02 Reverse DCF Solver

This notebook solves for assumptions implied by a target enterprise value. The result depends on the bracket and all other assumptions.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / "src")) if Path.cwd().name == "notebooks" else sys.path.insert(0, str(Path.cwd() / "src"))

from dataclasses import replace
from reversedcf.dcf import DCFInputs, enterprise_value
from reversedcf.metrics import format_percent
from reversedcf.reverse import solve_required_fcf_margin, solve_required_revenue_cagr
```

```python
base_inputs = DCFInputs(
    current_revenue=100_000_000_000,
    revenue_cagr=0.05,
    fcf_margin=0.20,
    forecast_years=5,
    wacc=0.09,
    terminal_growth=0.025,
    net_debt=0,
    shares_outstanding=1_000_000_000,
)

target_ev = enterprise_value(replace(base_inputs, revenue_cagr=0.12))
required_growth = solve_required_revenue_cagr(target_ev, base_inputs)
format_percent(required_growth)
```

```python
target_ev = enterprise_value(replace(base_inputs, fcf_margin=0.30))
required_margin = solve_required_fcf_margin(target_ev, base_inputs)
format_percent(required_margin)
```
