# 01 Basic DCF

This notebook runs a standard DCF using explicit assumptions. It is educational only and not financial advice.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent / "src")) if Path.cwd().name == "notebooks" else sys.path.insert(0, str(Path.cwd() / "src"))

from reversedcf.dcf import DCFInputs, run_dcf
from reversedcf.metrics import format_currency, format_percent
```

```python
inputs = DCFInputs(
    current_revenue=100_000_000_000,
    revenue_cagr=0.08,
    fcf_margin=0.22,
    forecast_years=5,
    wacc=0.09,
    terminal_growth=0.025,
    net_debt=5_000_000_000,
    shares_outstanding=1_000_000_000,
)

valuation = run_dcf(inputs)
format_currency(valuation.enterprise_value), format_currency(valuation.implied_share_price)
```
